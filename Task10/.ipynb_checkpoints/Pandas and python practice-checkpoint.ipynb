# Select rows with over 70,000 on the salary

```python
high_salary_rows = alumni[alumni['Salary'] > 70000]
high_salary_rows
```

# Update coulmn values in Marital Status to a charcheter crossponding it's value
for example Single ->S ,Married->M and so on

```python
marital_status_map = {
    'Single': 'S',
    'Married': 'M',
    'Divorced': 'D',
    'Widowed': 'W'
}

alumni['Marital Status'] = alumni['Marital Status'].map(marital_status_map)
alumni
```

# Drop the Fee coulmn

```python
alumni = alumni.drop(columns=['Fee'])
alumni 
```

# ADD a New row to the data set

```python
new_data = {
    'Name': 'Bassant',
    'Gender': 'Female',
    'Age': 21,
    'Marital Status': 'Single',
    'Salary': 10005,
    'year Graduated' : 2006
}
new_row_index = len(alumni)
alumni.loc[new_row_index] = new_data
alumni.tail()
```
