Now let's try make some subplots. A subplot is another name for a figure with multiple plots on it.

```python
# Create an empty subplot with 2 rows and 2 columns (4 subplots total)
import matplotlib.pyplot as plt

# Create an empty subplot with 2 rows and 2 columns (4 subplots total)
plt.figure(figsize=(10, 8))

# Creating subplots using subplot function
plt.subplot(2, 2, 1)
plt.title('Subplot 1')

plt.subplot(2, 2, 2)
plt.title('Subplot 2')

plt.subplot(2, 2, 3)
plt.title('Subplot 3')

plt.subplot(2, 2, 4)
plt.title('Subplot 4')

plt.tight_layout()  
plt.show()
```

Notice how the subplot has multiple figures. Now let's add data to each axes.

```python
# Create the same plot as above with 2 rows and 2 columns and figsize of (10, 5)
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(0)
X = np.linspace(0, 10, 100)
Y = X / 2
random_numbers = np.random.rand(10)
favorite_foods = {
    "Pizza": 10,
    "Sushi": 15,
    "Burger": 8
}
normally_distributed_data = np.random.randn(1000)


plt.figure(figsize=(10, 5))

# Plot X versus X/2 on the top left axes
plt.subplot(2, 2, 1)
plt.plot(X, Y)
plt.title('X vs X/2')

# Plot a scatter plot of 10 random numbers on each axis on the top right subplot

plt.subplot(2, 2, 2)
plt.scatter(random_numbers, random_numbers)
plt.title('Scatter Plot of 10 Random Numbers')
# Plot a bar graph of the favourite food keys and values on the bottom left subplot
plt.subplot(2, 2, 3)
plt.bar(favorite_foods.keys(), favorite_foods.values())
plt.title('Bar Graph of Favorite Foods')
plt.xlabel('Foods')
plt.ylabel('Prices')

# Plot a histogram of 1000 random normally distributed numbers on the bottom right subplot
plt.subplot(2, 2, 4)
plt.hist(normally_distributed_data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Normally Distributed Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()  
plt.show()
```

Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
import pandas as pd
from datetime import datetime
car_sales['Price'] = car_sales['Price'].replace('[\$,]', '', regex=True).astype(float)
print(car_sales.head())
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column

car_sales['Total Sales'] = car_sales['Price'].cumsum()
# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
today = datetime.today()
days = len(car_sales)
date_range = pd.date_range(today, periods=days)
car_sales['Sale Date'] = date_range
# View the car_sales DataFrame
print(car_sales.head())
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(car_sales['Sale Date'], car_sales['Total Sales'], marker='o', linestyle='-')
plt.title('Total Sales over Time')
plt.xlabel('Sale Date')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Convert the 'Price' column to the integers

car_sales['Price'] = car_sales['Price'].astype(int)
# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
plt.figure(figsize=(8, 6))
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.title('Scatter Plot of Odometer (KM) vs Price')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Importing heart-disease dataset
heart_disease = pd.read_csv('heart-disease.csv')

# Filtering patients over 50 years old
over_50 = heart_disease[heart_disease['age'] > 50]

# Defining colors for the target variable ('target'): blue for 0 (no heart disease), orange for 1 (heart disease)
colors = ['blue' if target == 0 else 'orange' for target in over_50['target']]

# Setting seaborn style to "whitegrid"
sns.set_style('whitegrid')

# Creating the plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(over_50['age'], over_50['chol'], c=colors, edgecolor='black', alpha=0.7)

# Adding labels and title
plt.title('Age vs Cholesterol for Patients over 50 (Heart Disease)')
plt.xlabel('Age')
plt.ylabel('Cholesterol (mg/dl)')

# Adding a meanline for cholesterol level
mean_chol = over_50['chol'].mean()
plt.axhline(y=mean_chol, color='red', linestyle='--', label=f'Mean Cholesterol ({mean_chol:.2f})')

# Adding legend
plt.legend(handles=scatter.legend_elements()[0], labels=['No Heart Disease', 'Heart Disease'], loc='upper right')

# Adjusting layout and showing grid
plt.tight_layout()
plt.grid(True)

# Showing the plot
plt.show()
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the `cmap` parameter of `scatter()` as well as the `color` parameter of `axhline()` to fix it.

Completing this step correctly should result in a figure which looks like this:
<img src="../images/matplotlib-heart-disease-chol-age-plot-cmap-change.png"/>

```python
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
scatter = plt.scatter(over_50['age'], over_50['chol'], c=over_50['target'], cmap='winter', edgecolor='black', alpha=0.7)
plt.title('Age vs Cholesterol for Patients over 50 (Heart Disease)')
plt.xlabel('Age')
plt.ylabel('Cholesterol (mg/dl)')
mean_chol = over_50['chol'].mean()
plt.axhline(y=mean_chol, color='red', linestyle='--', label=f'Mean Cholesterol ({mean_chol:.2f})')

# Adding legend
plt.legend(handles=scatter.legend_elements()[0], labels=['No Heart Disease', 'Heart Disease'], loc='upper right')

# Adjusting layout and showing grid
plt.tight_layout()
plt.grid(True)

# Showing the plot
plt.show()
```
